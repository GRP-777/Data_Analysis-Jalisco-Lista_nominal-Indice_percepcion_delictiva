```python
# Verificación de unidad en redacción de 'NOMBRE\nENTIDAD'
unique_entidades = df['NOMBRE\nENTIDAD'].unique()
print(len(unique_entidades))
print(unique_entidades)
```

```python
# Filtro por estado de 'JALISCO'
df_jalisco = df[df['NOMBRE\nENTIDAD'] == 'JALISCO']
```

```python
# Verificación de unidad en redacción de 'NOMBRE\nMUNICIPIO'
unique_entidades = df_jalisco['NOMBRE\nMUNICIPIO'].unique()
print(len(unique_entidades))
print(unique_entidades)
```

```python
# Agrupación por municipio, sumando las columnas de listas divididas por edad y género
columns_to_sum = [
    'LISTA\nNOMINAL', 'LISTA_18_HOMBRES', 'LISTA_18_MUJERES', 'LISTA_18_NOBINARIO',
    'LISTA_19_HOMBRES', 'LISTA_19_MUJERES', 'LISTA_19_NOBINARIO',
    'LISTA_20_24_HOMBRES', 'LISTA_20_24_MUJERES', 'LISTA_20_24_NOBINARIO',
    'LISTA_25_29_HOMBRES', 'LISTA_25_29_MUJERES', 'LISTA_25_29_NOBINARIO'
]

grouped_data = df_jalisco.groupby('NOMBRE\nMUNICIPIO')[columns_to_sum].sum()
df_municipios_jal = grouped_data
```

```python
# Creación de columnas totales, divididas por género
columns_to_sum = [
    'LISTA_18_HOMBRES', 'LISTA_19_HOMBRES', 
    'LISTA_20_24_HOMBRES', 'LISTA_25_29_HOMBRES'
]

df_municipios_jal['LISTA_HOMBRES_TOTAL'] = df_municipios_jal[columns_to_sum].sum(axis=1)

columns_to_sum = [
    'LISTA_18_MUJERES', 'LISTA_19_MUJERES', 
    'LISTA_20_24_MUJERES', 'LISTA_25_29_MUJERES'
]

df_municipios_jal['LISTA_MUJERES_TOTAL'] = df_municipios_jal[columns_to_sum].sum(axis=1)

columns_to_sum = [
    'LISTA_18_NOBINARIO', 'LISTA_19_NOBINARIO', 
    'LISTA_20_24_NOBINARIO', 'LISTA_25_29_NOBINARIO'
]

df_municipios_jal['LISTA_NOBINARIO_TOTAL'] = df_municipios_jal[columns_to_sum].sum(axis=1)
```

```python
# Creación de columna total, rango de edad 18-29
columns_to_sum = [
    'LISTA_NOBINARIO_TOTAL', 'LISTA_MUJERES_TOTAL', 'LISTA_HOMBRES_TOTAL'
]

df_municipios_jal['TOTAL_18_29'] = df_municipios_jal[columns_to_sum].sum(axis=1)

```

```python
# Conversión columnas tipo float a integer para prevenir mal funcionamiento
df_municipios_jal.fillna(0, inplace=True)
columns_to_convert = [
        'LISTA\nNOMINAL', 'TOTAL_18_29', 'LISTA_NOBINARIO_TOTAL',
       'LISTA_HOMBRES_TOTAL', 'LISTA_MUJERES_TOTAL', 'LISTA_18_HOMBRES',
       'LISTA_18_MUJERES', 'LISTA_18_NOBINARIO', 'LISTA_19_HOMBRES',
       'LISTA_19_MUJERES', 'LISTA_19_NOBINARIO', 'LISTA_20_24_HOMBRES',
       'LISTA_20_24_MUJERES', 'LISTA_20_24_NOBINARIO', 'LISTA_25_29_HOMBRES',
       'LISTA_25_29_MUJERES', 'LISTA_25_29_NOBINARIO'
]

df_municipios_jal[columns_to_convert] = df_municipios_jal[columns_to_convert].astype(int)
```

```python
# Modificación columna municipios para que sea rastreable geográficamente
df_municipios_jal["MUNICIPIO"] = df_municipios_jal.index
df_municipios_jal["MUNICIPIO"] = df_municipios_jal["MUNICIPIO"].apply(lambda x: x + ", JALISCO, MEXICO")
```

```python
# Creación de columnas de latitud y longitud para su manejo en Power BI

# Obtenga la clave de API de Google Maps
gmaps = googlemaps.Client(key="")


# Itere sobre los municipios
for municipio in df_municipios_jal["MUNICIPIO"]:

    # Geocodifique el municipio
    result = gmaps.geocode(municipio)

    # Obtenga las coordenadas geográficas
    lat = result[0].get("geometry").get("location").get("lat")
    lng = result[0].get("geometry").get("location").get("lng")

    # Agregue las coordenadas geográficas al DataFrame
    df_municipios_jal.loc[df_municipios_jal["MUNICIPIO"] == municipio, "LATITUD"] = lat
    df_municipios_jal.loc[df_municipios_jal["MUNICIPIO"] == municipio, "LONGITUD"] = lng
```

```python
# Reordenación de columnas
nuevo_orden = [
    'MUNICIPIO', 'LATITUD', 'LONGITUD',
    'LISTA\nNOMINAL', 'TOTAL_18_29', 'LISTA_NOBINARIO_TOTAL', 
    'LISTA_HOMBRES_TOTAL', 'LISTA_MUJERES_TOTAL', 
    'LISTA_18_HOMBRES', 'LISTA_18_MUJERES',
    'LISTA_18_NOBINARIO', 'LISTA_19_HOMBRES', 'LISTA_19_MUJERES',
    'LISTA_19_NOBINARIO', 'LISTA_20_24_HOMBRES', 'LISTA_20_24_MUJERES',
    'LISTA_20_24_NOBINARIO', 'LISTA_25_29_HOMBRES', 'LISTA_25_29_MUJERES',
    'LISTA_25_29_NOBINARIO'
]

df_municipios_jal = df_municipios_jal.reindex(columns=nuevo_orden)
```

```python
# Reemplazar 'BOLAÐOS' por 'BOLANIOS' en las columnas específicas
df_municipios_jal['NOMBRE\nMUNICIPIO'] = df_municipios_jal['NOMBRE\nMUNICIPIO'].str.replace('BOLAÐOS', 'BOLANIOS')
df_municipios_jal['MUNICIPIO'] = df_municipios_jal['MUNICIPIO'].str.replace('BOLAÐOS, JALISCO, MEXICO', 'BOLANIOS JALISCO, MEXICO')

df_municipios_jal['NOMBRE\nMUNICIPIO'] = df_municipios_jal['NOMBRE\nMUNICIPIO'].str.replace('ZAPOTLAN EL GRANDE', 'CIUDAD GUZMAN')
df_municipios_jal['MUNICIPIO'] = df_municipios_jal['MUNICIPIO'].str.replace('ZAPOTLAN EL GRANDE, JALISCO, MEXICO', 'CIUDAD GUZMAN, JALISCO, MEXICO')

df_municipios_jal['NOMBRE\nMUNICIPIO'] = df_municipios_jal['NOMBRE\nMUNICIPIO'].str.replace('TLAJOMULCO DE ZUÐIGA', 'TLAJOMULCO DE ZUNIGA')
df_municipios_jal['MUNICIPIO'] = df_municipios_jal['MUNICIPIO'].str.replace('TLAJOMULCO DE ZUÐIGA, JALISCO, MEXICO', 'TLAJOMULCO DE ZUNIGA, JALISCO, MEXICO')

df_municipios_jal['NOMBRE\nMUNICIPIO'] = df_municipios_jal['NOMBRE\nMUNICIPIO'].str.replace('CAÐADAS DE OBREGON', 'CANIADAS DE OBREGON')
df_municipios_jal['MUNICIPIO'] = df_municipios_jal['MUNICIPIO'].str.replace('CAÐADAS DE OBREGON, JALISCO, MEXICO', 'CANIADAS DE OBREGON, JALISCO, MEXICO')
```
